```python
import pandas as pd

# CSV dosyasını yükle
df = pd.read_csv('filmtv_movies.csv')  # Dosya adını kendi dosyana göre değiştir

# Kolon isimlerini göster
print("=" * 80)
print("📋 CSV KOLON İSİMLERİ:")
print("=" * 80)
print(df.columns.tolist())

# Toplam satır sayısı
print(f"\n📊 Toplam satır sayısı: {len(df)}")

# İlk 3 satırı göster
print("\n" + "=" * 80)
print("🎬 İLK 3 SATIR:")
print("=" * 80)
print(df.head(3))

# Her kolonun veri tipi ve boş olmayan değer sayısı
print("\n" + "=" * 80)
print("📈 KOLON BİLGİLERİ:")
print("=" * 80)
print(df.info())

# Eğer 'emotion' veya benzer kolon varsa, değerlerini göster
possible_emotion_cols = ['emotion', 'sentiment', 'feeling', 'mood', 'category']
for col in possible_emotion_cols:
    if col in df.columns:
        print(f"\n🎭 '{col}' kolonu değerleri:")
        print(df[col].value_counts())
```
