```python
import os
import pickle
import pandas as pd
import numpy as np
import faiss
from openai import OpenAI
from dotenv import load_dotenv
import warnings
warnings.filterwarnings('ignore')

from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS as LangChainFAISS
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain.schema import Document
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores.utils import DistanceStrategy

load_dotenv()
client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

print(" Kütüphaneler yüklendi!")
print(f"OpenAI API Key: {'Yüklendi' if os.getenv('OPENAI_API_KEY') else 'HATA!'}")
print("\n Dosya Durumu:")
print(f"  • movie_embeddings_full.pkl: {' VAR' if os.path.exists('movie_embeddings_full.pkl') else ' YOK - Cell 5 çalıştırın!'}")
print(f"  • langchain_faiss_db/: {' VAR' if os.path.exists('langchain_faiss_db') else ' YOK - İlk kez oluşturulacak'}")
```

```python
print("\n Embedding'ler yükleniyor...")
with open('movie_embeddings_full.pkl', 'rb') as f:
    data = pickle.load(f)
    df_sample = data['df']

print(f" {len(df_sample)} film yüklendi!")
print(f" Embedding boyutu: {len(df_sample['embedding'].iloc[0])}")
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores.utils import DistanceStrategy

class LangChainMovieRAG:
    """LangChain tabanlı Film Öneri RAG Sistemi"""
    
    def __init__(self, api_key):
        self.embeddings = OpenAIEmbeddings(
            model="text-embedding-3-small",
            openai_api_key=api_key
        )
        self.llm = ChatOpenAI(
            model="gpt-4o-mini",
            temperature=0.7,
            openai_api_key=api_key
        )
        self.vectorstore = None
        self.chain = None
        
    def create_vectorstore(self, df):
        """DataFrame'den FAISS vektör DB oluştur (batch processing)"""
        print("🚀 LangChain FAISS VectorStore oluşturuluyor...")
        print("💡 Mevcut embedding'ler kullanılıyor (hızlı yol)...")
        
        # Metinleri ve metadata'yı hazırla
        texts = []
        metadatas = []
        embeddings_list = []
        
        for idx, row in df.iterrows():
            text = f"Film: {row['title']}\nTür: {row['genre']}\nDuygu: {row['emotion']}\nYorum: {row['review'][:300]}"
            texts.append(text)
            metadatas.append({
                'title': row['title'],
                'genre': str(row['genre']),
                'emotion': row['emotion'],
                'review': row['review'][:200]
            })
            embeddings_list.append(row['embedding'])
        
        # FAISS'i embedding'lerle direkt oluştur (tekrar API çağrısı YOK!)
        embeddings_array = np.array(embeddings_list).astype('float32')
        
        # Manuel FAISS oluşturma
        dimension = len(embeddings_list[0])
        index = faiss.IndexFlatL2(dimension)
        index.add(embeddings_array)
        
        # Document objelerini oluştur
        documents = [
            Document(page_content=text, metadata=meta) 
            for text, meta in zip(texts, metadatas)
        ]
        
        # Docstore oluştur
        docstore = InMemoryDocstore({str(i): doc for i, doc in enumerate(documents)})
        index_to_docstore_id = {i: str(i) for i in range(len(documents))}
        
        self.vectorstore = LangChainFAISS(
            embedding_function=self.embeddings,
            index=index,
            docstore=docstore,
            index_to_docstore_id=index_to_docstore_id,
            distance_strategy=DistanceStrategy.EUCLIDEAN_DISTANCE
        )
        
        print(f" {len(documents)} doküman FAISS'e eklendi (API çağrısı YOK!)")
        return self.vectorstore
    
    def save_vectorstore(self, path="langchain_faiss_db"):
        if self.vectorstore:
            self.vectorstore.save_local(path)
            print(f" VectorStore kaydedildi: {path}")
    
    def load_vectorstore(self, path="langchain_faiss_db"):
        self.vectorstore = LangChainFAISS.load_local(
            path, 
            self.embeddings,
            allow_dangerous_deserialization=True
        )
        print(f" VectorStore yüklendi: {path}")
    
    def create_chain(self):
        if not self.vectorstore:
            raise ValueError("Önce vectorstore oluşturun!")
        
        memory = ConversationBufferMemory(
            memory_key="chat_history",
            return_messages=True,
            output_key="answer"
        )
     
        self.chain = ConversationalRetrievalChain.from_llm(
            llm=self.llm,
            retriever=self.vectorstore.as_retriever(
                search_type="mmr",  # Maximum Marginal Relevance - çeşitlilik sağlar
                search_kwargs={
                    "k": 10,  # 10 film getir
                    "fetch_k": 30,  # 30 adaydan seç
                    "lambda_mult": 0.5  # Çeşitlilik-benzerlik dengesi
                }
            ),
            memory=memory,
            return_source_documents=True,
            verbose=False
        )
        
        print("✅ LangChain RAG Chain hazır!")
        return self.chain
    
    def query(self, question):
        if not self.chain:
            self.create_chain()
        
        result = self.chain.invoke({"question": question})
        
        # Tekrarlayan filmleri temizle
        seen_titles = set()
        unique_docs = []
        
        for doc in result['source_documents']:
            title = doc.metadata.get('title', '')
            if title not in seen_titles:
                seen_titles.add(title)
                unique_docs.append(doc)
        
        return {
            'answer': result['answer'],
            'source_docs': unique_docs  # Tekrarsız filmler
        }

print("✅ LangChain RAG Class hazır!")
```

```python
api_key = os.getenv('OPENAI_API_KEY')
rag = LangChainMovieRAG(api_key)

vectorstore_path = "langchain_faiss_db"

if os.path.exists(vectorstore_path):
    print("📂 Mevcut FAISS DB yükleniyor...")
    rag.load_vectorstore(vectorstore_path)
else:
    print("🆕 Yeni FAISS DB oluşturuluyor (10-15 saniye)...")
    rag.create_vectorstore(df_sample)
    rag.save_vectorstore(vectorstore_path)

rag.create_chain()

print("\n" + "="*80)
print("✅ RAG SİSTEMİ HAZIR!")
print("="*80)
```

```python
def movie_chatbot():
    """LangChain tabanlı film öneri chatbot'u"""
    print("\n🎬 LANGCHAIN RAG FİLM ÖNERİ CHATBOTU")
    print("="*80)
    print("💡 Kullanılan Teknolojiler:")
    print("   • Generation Model: OpenAI GPT-4o-mini")
    print("   • Embedding Model: OpenAI text-embedding-3-small")
    print("   • Vector Database: FAISS")
    print("   • RAG Framework: LangChain")
    print("   • Retrieval: MMR (Maximum Marginal Relevance) - Çeşitli filmler")
    print("="*80)
    print("📝 Komutlar:")
    print("   • Film öneri iste: 'komik aksiyon filmi öner'")
    print("   • Çıkış: 'quit', 'exit', 'çık'")
    print("="*80)
    
    conversation_count = 0
    recommended_titles = set()  # Önerilen filmleri takip et
    
    while True:
        query = input("\n🎯 Siz: ").strip()
        
        if query.lower() in ['quit', 'exit', 'çık', 'q']:
            print("\n👋 Görüşmek üzere!")
            print(f"📊 Toplam {conversation_count} konuşma yapıldı.")
            print(f"🎬 {len(recommended_titles)} farklı film önerildi:")
            for title in list(recommended_titles)[:10]:
                print(f"   • {title}")
            break
        
        if not query:
            continue
        
        try:
            print("\n⏳ Düşünüyorum...")
            result = rag.query(query)
            
            # Önerilen filmleri kaydet
            for doc in result['source_docs']:
                recommended_titles.add(doc.metadata.get('title', ''))
            
            print(f"\n🤖 AI Asistanı:")
            print("-"*80)
            print(result['answer'])
            print("-"*80)
            
            print(f"\n📚 Kaynak filmler ({len(result['source_docs'])} farklı film):")
            for i, doc in enumerate(result['source_docs'][:5], 1):
                title = doc.metadata.get('title', 'Bilinmiyor')
                genre = doc.metadata.get('genre', '')
                emotion = doc.metadata.get('emotion', '')
                print(f"   {i}. 🎬 {title}")
                print(f"      Tür: {genre} | Duygu: {emotion}")
            
            conversation_count += 1
            
        except Exception as e:
            print(f"\n❌ Hata: {e}")
            print("💡 Lütfen tekrar deneyin.")
        
        print("\n" + "="*80)

print("✅ Chatbot fonksiyonu hazır!")
```

```python
movie_chatbot()
```
